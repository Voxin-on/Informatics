```python
import pandas as pd
workout=pd.read_excel("123.xlsx")
workout['Общая выручка'] = workout['Кол-во'] * workout['Стоимость']

svod_1 = workout.groupby(['Группа товаров', 'Название магазина'])['Общая выручка'].sum().unstack(fill_value=0)
print("Выручка по товарам у разных магазинов")
print(svod_1)

svod_2 = workout.groupby(['Группа товаров', 'Название магазина'])['Кол-во'].sum().unstack(fill_value=0)
print("\nКол-во проданных товаров у разных магазинов")
print(svod_2)

svod_3 = workout.groupby([workout['Дата продажи'].dt.month, 'Название магазина'])['Общая выручка'].sum().unstack(fill_value=0)
print("\nОбщая выручка разных магазинов по месяцам")
print(svod_3)

svod_4 = workout.groupby('Группа товаров')['Стоимость'].mean()
print("\nСредняя цена за штуку товара")
print(svod_4)

svod_5 = workout.groupby(workout['Дата продажи'].dt.month)['Кол-во'].sum()
print("\nОбщее количество продаж по месяцам")
print(svod_5)
```
